```python
import numpy as np

#Task: Perform the dot product of I and 9I+1 using numpy, here I is referred to as an 3x3 Identity matrix.

# Creating the 3x3 Identity matrix
I = np.identity(3)

# Creating the matrix 9I + 1
matrix_9I_plus_1 = 9 * I + 1

# Performing the dot product
result = np.dot(I, matrix_9I_plus_1)

print("Identity Matrix (I):\n", I)
print("\nMatrix (9I + 1):\n", matrix_9I_plus_1)
print("\nDot Product of I and (9I + 1):\n", result)
```

# Pandas Implementation

```python
import pandas as pd

## Create a DataFrame from a dictionary
data = {
    'Name': ['Ramesh', 'Mahesh', 'Suresh'],
    'Age': [25, 30, 35],
    'City': ['Bangalore', 'Mumbai', 'Delhi']
}
# Creating the DataFrame
df = pd.DataFrame(data)
print(df)

#Display the first 2 rows of the data frame
print(df.head(2))

#Print the age column
print(df['Age'])

#Filter rows where age is greater than 26
df= df[df['Age']>26]
print(df)

#Add a new column 'Country' with the value 'India' for all rows
df['Country']='India'
print(df)


data1 = {
    'Name': ['Ramesh', 'Mahesh', 'Suresh'],
    'Age': [25, None, 35],
    'City': ['Bangalore', 'Mumbai', 'Delhi']
}

df2= pd.DataFrame(data1)

# Fill missing values in the 'Age' column with the mean age
mean_age = df2['Age'].mean()
df2['Age'].fillna(mean_age, inplace=True)

# Display the DataFrame
print(df2)
```

# Matplotlib Implementation

```python
import matplotlib.pyplot as plt
import numpy as np     #numpy module. Fill your code

xpoints=np.array([1,2,3,4])
ypoints=np.array([2,4,6,8])

#Plot these points without drawing a line
plt.scatter(xpoints, ypoints)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of Points')
plt.show()

#Plotting with marker: Plot these points with a marker(Star marker)
plt.scatter(xpoints, ypoints, marker='*', color='blue', label='Data Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of Points with Star Marker')
plt.legend()
plt.show()

#Using fmt format, add circular marker,red color and Dashed line
plt.plot(xpoints, ypoints, 'ro--', label='Data Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Circular Marker, Red Color, and Dashed Line')
plt.legend()
plt.show()

#Add xlabel,ylabel and title for the plot.
plt.plot(xpoints, ypoints, 'ro--', label='Data Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Circular Marker, Red Color, and Dashed Line')
plt.legend()
plt.show()

#Create a scatter plot for xpoints and ypoints
plt.scatter(xpoints, ypoints, marker='o', color='blue', label='Data Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of Points')
plt.legend()
plt.show()

#Set color to the scatter plot. Blue,Green,Red and yellow color for each point respectively
colors = ['blue', 'green', 'red', 'yellow']
for x, y, color in zip(xpoints, ypoints, colors):
    plt.scatter(x, y, color=color, label=f'Point ({x}, {y})', s=100)  # s=100 sets marker size

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Different Colors')
plt.legend()
plt.show()
```

```python
import random
import numpy as np

#Set the seed of random to 20
np.random.seed(20)
random.seed(20)

arr1=np.array([1,24,31,45,73,81,94,25])

#Using the random module pick 4 different random numbers from arr1 and return their sum.
random_numbers = random.sample(list(arr1), 4)
sum_of_random_numbers = sum(random_numbers)
print(f"Random numbers picked: {random_numbers}")
print(f"Sum of random numbers: {sum_of_random_numbers}")
```
